```python
# Dependencies
import pandas as pd
```

```python
# read the csv files
measurements = pd.read_csv("Resources/hawaii_measurements.csv")
station = pd.read_csv("Resources/hawaii_stations.csv")
```

```python
# show 'hawaii_measurements.csv' dataframe
measurements.head()
```

```python
# show 'hawaii_stations.csv' dataframe
station.head()
```

```python
# Clean measurements dataframe 
# Inspect the data for NaNs and missing values.
pd.isnull(measurements)
```

```python
# Checking for null values
## alternate for checking null values: df.isnull().values.sum()
measurements.isnull().sum()
```

```python
# Removing the null values
measurements = measurements.dropna()
```

```python
measurements.head(5)
```

```python
# Save cleane dataframe
measurements.to_csv("Resources/clean_hawaii_measurements.csv", index=False)
```

```python
# Clean station datdframe
# Inspect the data for NaNs and missing values
pd.isnull(station)
```

```python
# Checking for null values
## alternate for checking null values: df.isnull().values.sum()
station.isnull().sum()
```

```python
# Removing the null values
station = station.dropna()
```

```python
station.head(5)
```

```python
# save clean dataframe
station.to_csv("Resources/clean_hawaii_stations.csv", index=False)
```
